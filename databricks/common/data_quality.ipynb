```python
from pyspark.sql import functions as F

def add_quality_flag(df, not_null_columns=[]):
    """ Add dq_flag column: VALID if all required columns not null, else INVALID """
    condition = F.lit(True)
    
    for col in not_null_columns:
        condition = condition & F.col(col).isNotNull()
    
    return df.withColumn(
        "dq_flag",
        F.when(condition, "VALID").otherwise("INVALID")
    )
```
