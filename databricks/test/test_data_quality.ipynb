```python
from pyspark.sql.functions import col

def test_add_quality_flag_with_null():
    """Test record marked INVALID when required column is null"""
    data = [
        ("A123", "490001", "25"),
        ("A124", None, "38")
    ]
    df = spark.createDataFrame(data, ["arrival_id", "naptan_id", "line_id"])
    
    result = add_quality_flag(df, not_null_columns=["arrival_id", "naptan_id", "line_id"])
    
    assert result.filter(col("dq_flag") == "VALID").count() == 1
    assert result.filter(col("dq_flag") == "INVALID").count() == 1
    print("âœ… test_add_quality_flag_with_null PASSED")

test_add_quality_flag_with_null()
```
