```python
from pyspark.sql.functions import col

def test_trim_strings():
    data = [("  hello  ", 1), ("world", 2)]
    df = spark.createDataFrame(data, ["name", "id"])
    
    result = trim_strings(df)
    
    assert result.filter(col("name") == "hello").count() == 1
    print("✅ test_trim_strings PASSED")


def test_add_transformation_metadata():
    data = [("A123", "Station1")]
    df = spark.createDataFrame(data, ["id", "station"])
    
    result = add_transformation_metadata(df)
    
    assert "transformation_timestamp" in result.columns  # ✅ Built-in assert
    print("✅ test_add_transformation_metadata PASSED")


# Run all tests
test_trim_strings()
test_add_transformation_metadata()
```
